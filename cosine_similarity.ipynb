```python
!pip install numpy opencv-python # Install the necessary libraries
import cv2 # Import the cv2 module
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity

def load_img(img_path):
  image = cv2.imread(img_path)
  image = cv2.resize(image,(128,128))
  image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
  image = image.flatten()
  return image


img_1 = load_img('/content/img_1.jpg')
img_2 = load_img('/content/img_2.jpg')
```

```python
similarity = cosine_similarity([img_1, img_2])
print(similarity)
```
